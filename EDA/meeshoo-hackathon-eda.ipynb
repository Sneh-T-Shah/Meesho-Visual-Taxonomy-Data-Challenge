```python
def specific_null_column(df, category_name):
    null_cases = {}
    total_nulls = 0  # Variable to accumulate the total count of null values

    # Iterate through each unique 'len' value
    for length in df['len'].unique():
        # Filter rows where 'len' matches the current length
        df_filtered = df[df['len'] == length]
        
        # Select only the relevant attributes up to the length specified
        attr_columns = [f'attr_{i+1}' for i in range(length)]
        null_counts = df_filtered[attr_columns].isnull().sum(axis=1)
        
        # Loop through each possible null count from 1 to the specified length
        for i in range(1, length + 1):
            # Count rows with exactly 'i' NaNs, excluding any rows with more than 'i' NaNs
            count = (null_counts == i).sum()
            null_cases[f"{i} column(s) null (for len={length})"] = count
            total_nulls += count  # Add count to total_nulls
    
    # Display results
    for key, value in null_cases.items():
        print(f"{key}: {value}")
    
    # Print the total sum of values
    print(f"Total sum of NaN cases: {total_nulls}")
    print("------------------------------\n")
```

```python
df_c1 = df[df['Category'] == 'Men Tshirts']
df_c2 = df[df['Category'] == 'Sarees']
df_c3 = df[df['Category'] == 'Kurtis']
df_c4 = df[df['Category'] == 'Women Tshirts']
df_c5 = df[df['Category'] == 'Women Tops & Tunics']
```

```python
print(f"length of the df_c1 before removing the NaN rows and pre-processing: {len(df_c1)}")
specific_null_column(df_c1, "c1")

print(f"length of the df_c2 before removing the NaN rows and pre-processing: {len(df_c2)}")
specific_null_column(df_c2, "c2")

print(f"length of the df_c3 before removing the NaN rows and pre-processing: {len(df_c3)}")
specific_null_column(df_c3, "c3")

print(f"length of the df_c4 before removing the NaN rows and pre-processing: {len(df_c4)}")
specific_null_column(df_c4, "c4")

print(f"length of the df_c5 before removing the NaN rows and pre-processing: {len(df_c5)}")
specific_null_column(df_c5, "c5")
```
